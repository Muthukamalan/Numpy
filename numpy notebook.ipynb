<h1 align="center"> Blurr Image Exercise</h1>

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = plt.imread('./dc_metro.png')
print("shape: ",img.shape)
print("dtype: ",img.dtype)
print(img)
```

```python
img_row=img.shape[0]
img_col=img.shape[1]

blur_img = np.zeros((img_row-2,img_col-2))
for i in range(1,img_row-1):
    for j in range(1,img_col-1):
        left  = img[i,j-1]
        bottom= img[i+1,j]
        top   = img[i-1,j]
        center= img[i,j]
        right = img[i,j+1]
        
        mean_ = (left+right+top+bottom+center)/5
        blur_img[i-1,j-1]=mean_

print(" blur image: ", blur_img.shape)

difference_img = img- np.pad(blur_img,1,'constant')
print("difference: ",difference_img.shape)
```

```python
fig,(ax1,ax2,ax3)= plt.subplots(1,3)

ax1.imshow(img,cmap='gray')
ax1.axis("Off")

ax2.imshow(blur_img,cmap='gray')
ax2.axis("Off")

ax3.imshow(difference_img,cmap='gray')
ax3.axis("Off")

plt.show()
```

```python
def blur_function(image_):
    original_img=image_.copy()
    img_row=image_.shape[0]
    img_col=image_.shape[1]

    blurred_img = np.zeros((img_row-2,img_col-2))
    for i in range(1,img_row-1):
        for j in range(1,img_col-1):
            left  = image_[i,j-1]
            bottom= image_[i+1,j]
            top   = image_[i-1,j]
            center= image_[i,j]
            right = image_[i,j+1]
        
            mean_ = (left+right+top+bottom+center)/5
            blurred_img[i-1,j-1]=mean_
            del mean_
            
    image_ = blurred_img
    
    del img_row,img_col,blurred_img
    
    fig,(ax1,ax2,ax3)= plt.subplots(1,3)

    ax1.imshow(image_,cmap='gray')
    ax1.axis("Off")

    ax2.imshow(original_img,cmap='gray')
    ax2.axis("Off")
    
    pixel_difference = (original_img.shape[0]-image_.shape[0])//2
    difference_img   =  original_img - np.pad(image_,pixel_difference,'constant')
    
    ax3.imshow(difference_img,cmap='gray')
    ax3.axis("Off")

    plt.show()
```

```python
blur_function(image_=img)
```

```python
def blur_50_times(image_):
    original_img = image_
    for _ in range(50):
        img_row=image_.shape[0]
        img_col=image_.shape[1]

        blurred_img = np.zeros((img_row-2,img_col-2))
        for i in range(1,img_row-1):
            for j in range(1,img_col-1):
                left  = image_[i,j-1]
                bottom= image_[i+1,j]
                top   = image_[i-1,j]
                center= image_[i,j]
                right = image_[i,j+1]
            
                mean_ = (left+right+top+bottom+center)/5
                blurred_img[i-1,j-1]=mean_
            
        image_ = blurred_img
    return image_
```

```python
img = plt.imread('./dc_metro.png')
blurred_image = blur_50_times(image_=img)

# PLOT
plt.figure(figsize=(15,15))
fig,(ax1,ax2,ax3)= plt.subplots(1,3)
ax1.imshow(blurred_image,cmap='gray')
ax1.axis("Off")

ax2.imshow(img,cmap='gray')
ax2.axis("Off")
    
pixel_difference = (img.shape[0]-blurred_image.shape[0])//2
difference_img   =  img - np.pad(blurred_image,pixel_difference,'constant')
    
ax3.imshow(difference_img,cmap='gray')
ax3.axis("Off")

plt.show()
```
